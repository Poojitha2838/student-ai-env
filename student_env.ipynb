```python
class StudentEnv:
    def __init__(self, level="easy"):
        self.level = level
        self.reset()

    def reset(self):
        self.marks = 0
        self.steps = 0
        return self.state()

    def step(self, action):
        self.steps += 1

        if action == "study":
            reward = 10
        elif action == "practice":
            reward = 15
        elif action == "skip":
            reward = -5
        else:
            reward = 0

        self.marks += reward

        done = False
        max_marks = 50 if self.level == "easy" else 100 if self.level == "medium" else 200

        if self.marks >= max_marks or self.steps >= 10:
            done = True

        return self.state(), reward, done

    def state(self):
        return {"marks": self.marks, "steps": self.steps}
```

```python
env = StudentEnv(level="easy")

state = env.reset()
print("Start:", state)

for i in range(5):
    state, reward, done = env.step("study")
    print("Step:", state, "Reward:", reward)

    if done:
        break
```
